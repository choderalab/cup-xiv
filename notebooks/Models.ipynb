```python
import numpy as np
from pylab import *

nequil = 500 # number of samples to discard to equilibration

# Load MCMC trace of experimental error for GBSA parameter fit.
sigma_t = np.genfromtxt('data/MCMC.txt/Chain_0/sigma.txt',dtype='float')[nequil:]

# Plot histogram of combined error.
rcParams['figure.figsize'] = [10, 5];
hist(sigma_t);
xlabel('model and experimental combined error (kcal/mol)');
ylabel('P(error)');
yticks([]);
print 'combined error estimate is %.2f +- %.2f kcal/mol' % (sigma_t.mean(), sigma_t.std())
```

```python
# Load MCMC trace of deviance.
deviance_t = np.genfromtxt('data/MCMC.txt/Chain_0/deviance.txt',dtype='float')[nequil:]

# Plot trace.
rcParams['figure.figsize'] = [10, 5];
plot(deviance_t, 'o');
xlabel('MCMC iteration');
ylabel('model deviance');
```

```python
# Evaluation
evaluate = np.genfromtxt('data/gbsa-evaluate.txt',dtype='float')[nequil:,:]
```

```python
compound_index = 20
hist(evaluate[:,compound_index]);
```
